```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, LogisticRegression
```

```python
data = pd.read_csv('Student_Performance_Dataset_2024.csv', delimiter=',')
data
```

```python
X = data[['Quiz', 'Midterm','Categories']].values
X
```

```python
# 2. Créer Pass_Status (ATTENTION: on utilise .strip() au cas où il y a des espaces cachés dans le mot 'Fail')
data['Pass_Status'] = data['Categories'].astype(str).str.strip().apply(lambda x: 0 if x == 'Fail' else 1)

# 3. Séparer X et y
X = data[['Quiz', 'Midterm']].values
y = data[['Pass_Status']].values.ravel()
```

```python
# plot
ind0=np.nonzero(y==0)[0]
ind1=np.nonzero(y==1)[0]
plt.scatter(X[ind0,0],X[ind0,1], s=30, c='y', marker='o', linewidths=1, label='Fail')
plt.scatter(X[ind1,0],X[ind1,1], s=30, c='b', marker='+', linewidths=1, label='Pass')
plt.xlabel('Quiz')
plt.ylabel('Midterm')
plt.legend()
```

```python
### 3. Build the models from scratch
# --- Logistic Regression ---
```

```python
def sigmoid(z):
    sig = 1/(1+np.exp(-z))
    return sig
```

```python
sigmoid(0)
```

```python
def costFunction(X, y, theta=[[0], [0], [0]]):
    m = y.size
    X_extend = np.c_[np.ones(m),X]
    z = X_extend.dot(theta)
    h = sigmoid(z)
    
    epsilon = 1e-15
    h = np.clip(h, epsilon, 1 - epsilon)
    
    cost = (-1/m)*np.sum(y*np.log(h)+(1-y)*np.log(1-h))
    
    return cost
```

```python
costFunction( X , y )
```

```python
def gradientDescent(X, y, theta=[[0], [0], [0]], epochs = 1000, learning_rate = 0.01):
    m = y.size
    X_extend = np.c_[np.ones(m), X]
    j_history = np.zeros(epochs)
    n_print = epochs//10

    for i in range(epochs):
        z = X_extend.dot(theta)
        h = sigmoid(z)
        epsilon = 1e-15
        h = np.clip(h, epsilon, 1 - epsilon)
        
        theta[0] = theta[0]-(learning_rate/m)*np.sum(h-y)
        theta[1] = theta[1]-(learning_rate/m)*np.sum((h-y)*X[:,0:1])
        theta[2] = theta[2]-(learning_rate/m)*np.sum((h-y)*X[:,1:2])
       
        j = (-1/m)*np.sum(y*np.log(h)+(1-y)*np.log(1-h))
        j_history[i] = j
        
        if (i+1) % n_print == 0:
            print("Iteration", i+1, " | Cost: ", j, "\n")
    
    return  theta, j_history
```

```python
X_norm = (X - np.mean(X, axis=0)) / np.std(X, axis=0)
theta, j_history = gradientDescent(X_norm, y, epochs=5000, learning_rate=0.1)
```

```python
theta
```

```python
plt.plot( j_history )
plt.xlabel('Epochs')
plt.ylabel('Cost value')
```

```python
pos = (y == 1).ravel()
neg = (y == 0).ravel()
```

```python
x_plot = np.array((X[:, 1].min(), X[:, 1].max()))
y_plot = -(1/theta[2]) * (theta[0] + theta[1] * x_plot)

plt.plot(X[pos, 0], X[pos, 1], 'k+', X[neg, 0], X[neg, 1], 'yo')
plt.plot(x_plot, y_plot)
plt.xlabel('Quiz')
plt.ylabel('Midterm')
plt.legend(loc='upper center', labels=['Pass', 'Not pass', 'Decision boundary'], bbox_to_anchor=(0.5,-0.15));
```

```python
def predict(X, theta):
    X_extend=np.c_[1,X]
    z = X_extend.dot(theta)
    h = sigmoid(z)
    if h>=0.5: 
        return 1
    else:
        return 0
```

```python
X_sample = np.array([[25, 80]])

result = predict(X_sample, theta)
print("Prediction:", result)
```
